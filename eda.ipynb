```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = pd.read_csv('reduced_version_data_ENEL_645.csv')
data
```

```python
fig, ax = plt.subplots(1,3)
cols = ['Resident Count', 'Year', 'Category']
for i in range(len(ax)):
    feature = cols[i]
    ax[i].scatter(data[feature],data["Crime Count"], label=feature)
    ax[i].set_xlabel(feature)
    ax[i].legend()
ax[0].set_ylabel("Crime Count")
plt.show()
```
